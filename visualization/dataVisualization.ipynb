```python
import matplotlib.pyplot as plt
import pandas as pd
plt.rcParams['figure.figsize'] = [10, 6]

sit = pd.read_csv("/content/drive/MyDrive/fall_detection_dataset/ADL/sit/sitchair.csv")
stand = pd.read_csv("/content/drive/MyDrive/fall_detection_dataset/ADL/stand/stand.csv")
walk = pd.read_csv("/content/drive/MyDrive/fall_detection_dataset/ADL/walk/walk.csv")
forward_bend = pd.read_csv("/content/drive/MyDrive/fall_detection_dataset/ADL/forward_bend/forwardBend.csv")
forward_bend_deep = pd.read_csv("/content/drive/MyDrive/fall_detection_dataset/ADL/forward_bend_deep/forwardBendDeep.csv")


fall_forward = pd.read_csv("/content/drive/MyDrive/fall_detection_dataset/fall/fall_forward/fall.csv")
fall_backward = pd.read_csv("/content/drive/MyDrive/fall_detection_dataset/fall/fall_backward/fallBackward.csv")
fall_leftside = pd.read_csv("/content/drive/MyDrive/fall_detection_dataset/fall/fall_leftside/fallside1.csv")

# sit_data.info()
sit.head(10)
# sit_data.isnull().any()
```

# Accelerometer Data

```python

%matplotlib inline
plt.figure(figsize=(10,6))

def plotAcceleration(df, title):
  # plt.xlim()
  plt.ylim(-25, 25)
  ax = df['ax']
  ay = df['ay']
  az = df['az']

  plt.plot(ax, label='ax')
  plt.plot(ay, label ='ay')
  plt.plot(az, label='az')
  plt.xlabel('Sample', fontsize=14)
  plt.ylabel('Acceleration (m/s)', fontsize=14)
  plt.title(title, fontsize=18)
  plt.grid()
  plt.legend()
  plt.show()
  plt.figure(figsize=(10,6))

plotAcceleration(sit, "Sitting")
plotAcceleration(stand, "Standing")
plotAcceleration(walk, "Walking")
plotAcceleration(forward_bend, "Forward Bend")
plotAcceleration(forward_bend_deep, "Forward Bend Deep")

plotAcceleration(fall_forward, "Forward Fall")
plotAcceleration(fall_backward, "Backward Fall")
plotAcceleration(fall_leftside, "Fall to the Left")
```

```python
import numpy as np
plt.figure(figsize=(10,6))

def plotAccelerationTotal(df, title):
  plt.ylim(0, 35)
  ax = df['ax']
  ay = df['ay']
  az = df['az']
  a = np.sqrt(ax**2 + ay**2 + az**2)
  plt.plot(a, label='a')
  plt.xlabel('Sample', fontsize=14)
  plt.ylabel('Acceleration (m/s)', fontsize=14)
  plt.title(title, fontsize=18)
  plt.grid()
  plt.legend()
  plt.show()
  plt.figure(figsize=(10,6))

plotAccelerationTotal(sit, "Sitting")
plotAccelerationTotal(stand, "Standing")
plotAccelerationTotal(walk, "Walking")
plotAccelerationTotal(forward_bend, "Forward Bend")
plotAccelerationTotal(forward_bend_deep, "Forward Bend Deep")

plotAccelerationTotal(fall_forward, "Forward Fall")
plotAccelerationTotal(fall_backward, "Backward Fall")
plotAccelerationTotal(fall_leftside, "Fall to the Left")
```

# Gyroscope Data

```python
%matplotlib inline
plt.figure(figsize=(10,6))

def plotGyro(df, title):
  # plt.xlim()
  plt.ylim(-250, 300)
  ax = df['gx']
  ay = df['gy']
  az = df['gz']

  plt.plot(ax, label='gx')
  plt.plot(ay, label ='gy')
  plt.plot(az, label='gz')
  plt.xlabel('Sample', fontsize=14)
  plt.ylabel('Angular Velocity (degree/s)', fontsize=14)
  plt.title(title, fontsize=18)
  plt.grid()
  plt.legend()
  plt.show()
  plt.figure(figsize=(10,6))

plotGyro(sit, "Sitting")
plotGyro(stand, "Standing")
plotGyro(walk, "Walking")
plotGyro(forward_bend, "Forward Bend")
plotGyro(forward_bend_deep, "Forward Bend Deep")

plotGyro(fall_forward, "Forward Fall")
plotGyro(fall_backward, "Backward Fall")
plotGyro(fall_leftside, "Fall to the Left")
```

```python
plt.figure(figsize=(10,6))


def plot_gyro_total(df, title):
  plt.ylim(0, 350)
  gx = df['gx']
  gy = df['gy']
  gz = df['gz']
  
  g = np.sqrt(gx**2 + gy**2 + gz**2)
  
  plt.plot(g, label='Angular Velocity (degree/s)')
  plt.xlabel('Sample', fontsize=14)
  plt.ylabel('Angular Velocity (degree/s)', fontsize=14)
  plt.title(title, fontsize=18)
  plt.grid()
  plt.legend()
  plt.show()
  plt.figure(figsize=(10,6))

plot_gyro_total(sit, "Sitting")
plot_gyro_total(stand, "Standing")
plot_gyro_total(walk, "Walking")
plot_gyro_total(forward_bend, "Forward Bend")
plot_gyro_total(forward_bend_deep, "Forward Bend Deep")

plot_gyro_total(fall_forward, "Forward Fall")
plot_gyro_total(fall_backward, "Backward Fall")
plot_gyro_total(fall_leftside, "Fall to the Left")
```
